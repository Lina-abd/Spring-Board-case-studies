```python
#Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members!

%%sql 

SELECT name, SUM(totalrevenue) AS "TotalRevenue"
FROM (
    SELECT Bookings.memid,
           Bookings.facid,
           Facilities.membercost,
           Facilities.guestcost,
           Facilities.name,
           Bookings.slots,
           Bookings.starttime,
           CASE WHEN Bookings.memid = 0 THEN Facilities.guestcost * Bookings.slots
                WHEN Bookings.memid > 0 THEN Facilities.membercost * Bookings.slots
           END AS totalrevenue
    FROM Bookings
    JOIN Facilities ON Bookings.facid = Facilities.facid
) AS Table1

GROUP BY name

HAVING SUM(totalrevenue) < 1000

ORDER BY SUM(totalrevenue);
```

```python
#Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order
```

```python
%%sql

SELECT
    m1.memid AS member_id,
    m1.firstname AS member_firstname,
    m1.surname AS member_surname,
    m1.recommendedby AS recommended_by_id,
    m2.firstname AS recommended_by_firstname,
    m2.surname AS recommended_by_surname
FROM
    Members AS m1
LEFT JOIN
    Members AS m2 ON m1.recommendedby = m2.memid
ORDER BY
    m1.surname,
    m1.firstname;
```

```python
#Q12: Find the facilities with their usage by member, but not guests
```

```python
%%sql

SELECT t1.memid, t1.surname || ' ' || t1.firstname AS MemberName, t2.name, t2.FacUsageSlots, t2.MembersUsageCost
FROM Members AS t1
JOIN (
    SELECT Facilities.facid, Bookings.memid, Facilities.name,
    SUM(Bookings.slots) AS "FacUsageSlots", Facilities.membercost *
    SUM(Bookings.slots) AS "MembersUsageCost"
    FROM Bookings
    JOIN Facilities ON Bookings.facid = Facilities.facid
    WHERE Bookings.memid > 0
    GROUP BY Bookings.memid
    ) AS t2
ON t1.memid = t2.memid
ORDER BY t2.MembersUsageCost;
```

```python
#Q13: Find the facilities usage by month, but not guests
```

```python
%%sql

SELECT
    strftime('%m', starttime) AS booking_month,
    Facilities.facid,
    Facilities.name,
    COUNT(*) AS bookings_count
FROM
    Bookings
JOIN
    Facilities ON Bookings.facid = Facilities.facid
WHERE Bookings.memid > 0
GROUP BY
    booking_month, Facilities.facid, Facilities.name
ORDER BY
    booking_month, Facilities.facid;
```
